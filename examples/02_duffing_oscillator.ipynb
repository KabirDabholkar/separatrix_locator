# Tutorial 2: Duffing Oscillator

This tutorial demonstrates separatrix location in a 2D dynamical system - the Duffing oscillator.

The system is defined by:
$$
\frac{dx}{dt} = y
$$
$$
\frac{dy}{dt} = -y + x - x^3
$$

This system has:
- Two stable fixed points at $(\pm 1, 0)$
- One unstable fixed point (separatrix) at $(0, 0)$
- Damping that creates spiral behavior

## Learning Objectives
- Work with 2D dynamical systems
- Visualize phase portraits and trajectories
- Understand how separatrix location works in higher dimensions
- Compare different model architectures
